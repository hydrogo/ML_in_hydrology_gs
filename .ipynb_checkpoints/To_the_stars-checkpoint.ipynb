# Storming

```python
# correlations of course
raw_data.corr()
```

```python
#what about mounthly reshaping (aggregation)?
# try to use .resample method
Qobs_mon = raw_data.Qobs.resample("M", how='median')
Qobs_mon.plot()
```

```python
Temp_mon = raw_data.Temp.resample("M", how='median')
Temp_mon.plot()
```

```python
Ptot_mon = raw_data.Ptot.resample("M", how='sum')
Ptot_mon.plot()
```

```python
raw_data_mon = pd.concat([Qobs_mon, Temp_mon, Ptot_mon], axis=1)
raw_data_mon.head()
```

```python
raw_data_mon.corr()
```

```python
raw_data_mon[['Qobs', 'Ptot']].plot(kind='scatter', x='Qobs', y='Ptot')
```

```python
#let's get some year aggregation, why not!
Qobs_year = raw_data.Qobs.resample("A", how='median')
Temp_year = raw_data.Temp.resample("A", how='median')
Ptot_year = raw_data.Ptot.resample("A", how='sum')
raw_data_year = pd.concat([Qobs_year, Temp_year, Ptot_year], axis=1)
raw_data_year.corr()
```

```python
raw_data_year[['Qobs', 'Ptot']].plot(kind='scatter', x='Qobs', y='Ptot')
```

**It was some hypotheses test. And conclusion is: the smaller time lag the higher complexity research level (and uncertainties)**

##**Hypotheses forming**

The most simple way is 
Q = f(P, T, PE)

Learn model on 1980 - 2010 period (31 year). Check model accuracy on 2011 year.

Ok, no problem. Let's build the model!

### Step 1. Feature ingeneering
* The most important part
* Skip it

### Step 2. Preprocessing
* Normalization, Standartization
* Dataset separation

```python
#divide our data
train_raw = raw_data['1980':'2010']
test_raw = raw_data['2011']
```

```python
#train_raw.describe()
#test_raw.describe()
```

```python
# form vector of features
X = train_raw[['Ptot', 'Temp', 'PE']].values

# and target vector too
y = np.ravel(train_raw[['Qobs']].values)

print(X.shape, y.shape)
```

Important thing is that all features been related to the [0; 1] interval

```python
'''#let's import some scikit-learn library module for normalizing
from sklearn import preprocessing
X = preprocessing.normalize(X)
y = np.ravel(preprocessing.normalize(y))'''
```

```python
#do the modelling
from sklearn import ensemble
from sklearn.metrics import mean_squared_error

# set the parameters of our ML model

'''params = {'n_estimators': 500, 'max_depth': 4, 'min_samples_split': 1,
          'learning_rate': 0.01, 'loss': 'ls'} '''

# init the model 
# we'll use Gradient Boosting Regressor, why not?
model = ensemble.GradientBoostingRegressor()
```

```python
# learn the model ("fit" in terms of library)
model.fit(X, y)
```

```python
# prepare examination dataset
X_test = test_raw[['Ptot', 'Temp', 'PE']].values
y_test = np.ravel(test_raw[['Qobs']].values)
```

```python
mse = mean_squared_error(y_test, model.predict(X_test))
print(mse)
```

```python
results = pd.DataFrame({'modelled': model.predict(X_test), \
                       'expected': y_test})
```

```python
results.plot()
```

```python
results.corr()
```

**It was one another hypotheses test. 
And conclusion is: the smarter you the better result you receive**
